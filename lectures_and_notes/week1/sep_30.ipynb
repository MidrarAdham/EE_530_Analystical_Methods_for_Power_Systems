## Background

#### Power System Main Components:
- Generation
- Transmission
- Distribution
- Consumption

These components can be viewed as several hierarchical levels. Each level encompasses different range of voltages.

- FERC & NERC oversee the bulk power system, ranging from generation to consumption.
- While FERC's main goal is to regulate the interstate transmission systems, NERC focuses on maintaining the reliability and security of the grid. [(Know more!)](https://www.ferc.gov/reliability-explainer)
- Note that the power flows from the source (Generation) to the load (consumers)
- We will cover the rest of these entities later in this document.

**The current status of the grid**
- The power system is going through fundamental changes.
- These changes mainly impact the generation, distribution, and consumption.
- Transmission system is also impacted, but not as much as the other ones.
- These changes can be summarized as follows:
    - Generation is no longer centralized. Power can be generated from the consumer side.
    - Distribution systems are experiencing heavy demand (i.e. power consumption), prompting the need for equipment upgrades, automation, etc.
    - Many of the loads are now grid-interactive; they can operate in an automated manner.

**Because of these factors, new components are introduced to the power system. See below**

```python
SVG(filename='fig2.svg')
```

**What changed?**
- Power flow is now bi-directional.
- There is a new entity here: **Aggregators**.
- Also, consumers are now "prosumers"

**Aggregators:**
- This is a vague, new concept that's been introduced to the system.
- The definition of aggregators is clear, however, its responsibilities are not clearly defined.
- Aggregators are either business entities or utility-related entities that aggregate customer loads.
- Aggregators then advertise these aggregate loads to utilities, so they can be controlled to ensure the reliability of the grid.
